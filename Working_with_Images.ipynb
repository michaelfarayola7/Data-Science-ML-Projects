# Working with an image
NB: The image is a free to downloa image from pexels.com


```python
import skimage.io as io

import matplotlib.pyplot as plt
import numpy as np
```

Importing the image

```python
img = io.imread("/content/Girl.jpeg.jpg")
plt.imshow(img)
plt.show()
```

```python
img.shape
```

Extracting all the channels of the image

```python
red = img[:,:,0]
green = img[:,:,1]
Blue = img[:,:,2]
```

```python
plt.imshow(red, cmap="Reds")
plt.show()
```

```python
plt.imshow(green, cmap="Greens")
plt.show()
```

```python
plt.imshow(Blue,cmap="Blues")
plt.show()
```

#Affine Transformation

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import apply_affine_transform
```

```python
transformation = apply_affine_transform(
    img,
    theta=270
)
```

```python
plt.imshow(transformation)
plt.show()
```

```python
# Read the image
img3 = io.imread('/content/girl2.jpeg')

plt.imshow(img3)
plt.show()
```

```python
import tensorflow as tf

transformation = tf.keras.preprocessing.image.apply_affine_transform(
img3,
theta=270
)

plt.imshow(transformation.astype('uint8'))
plt.axis('off')  # To hide the axis values
plt.show()
```

```python
#zx: zoom in x direction
#zy: zoom in y direction

transformation = apply_affine_transform(img, zx=0.15, zy=0.8)
plt.imshow(transformation)
plt.show()
```

```python
# tx: Width shift.
# ty: Height shift.

transformation = tf.keras.preprocessing.image.apply_affine_transform(img, tx=50, ty=1)
plt.imshow(transformation)
```
